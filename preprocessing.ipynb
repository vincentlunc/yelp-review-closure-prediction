```python
'''
0330
goal: balance closed and unclosed restaurant
'''
import pandas as pd
import numpy as np
busi = pd.read_table('restaurant_balanced.txt')
```

```python
len(busi[busi['is_open']==0])
```

```python
is_closed = busi[busi['is_open'] == 0]
is_open = busi[busi['is_open'] == 1].sample(len(is_closed))
busi_balance = is_closed.append(is_open, ignore_index=True)
len(busi_balance)
```

```python
busi_balance.tail()
```

```python
# apply to review
review = pd.read_table('restaurant_review.txt')
business_id = busi_balance[['business_id']]
out = review.merge(business_id, on='business_id')
```
